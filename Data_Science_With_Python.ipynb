```python
#convert this string into numerical representations (Encoding the string or state column)


pd.get_dummies(df)

```

```python
#here we need to drop the first cause if other two is zero then the this indicates the first one

dataf= pd.get_dummies(df,drop_first= True)
```

```python
dataf.head()
```

#Split the data into train and test

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score, mean_squared_error
```

```python
dataf.columns
```

```python
#define the variables

features=['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']

target= ['Profit']

X = dataf[features]
y= dataf[target]
```

```python
#Training and test set

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state = 21)
```

```python
lin= LinearRegression()
lin.fit(X_train, y_train)
y_pred= lin.predict(X_test)
y_pred
```

```python
r2_score(y_test, y_pred)
```

```python
#insert  prediction column

dataf.insert(4,"Predictions",lin.predict(X))
```

```python
#save as csv

dataf.to_csv('full_dataset.csv')
```
